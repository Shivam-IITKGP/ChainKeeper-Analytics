```python
#import pages data
pages =  pd.read_csv('page.csv',nrows=550)
```

```python
#explore the data types of the dataframe
pages.dtypes
```

```python
#remove null values
pages =  pages.fillna('-1')
```

```python
#explore data types of dataframe
pages.columns
```

```python
#encode the data
label_encoder = preprocessing.LabelEncoder()
pages['url'] = label_encoder.fit_transform(pages['url'])
pages['title'] = label_encoder.fit_transform(pages['title'])
pages['code'] = label_encoder.fit_transform(pages['code'])
pages['domain'] = label_encoder.fit_transform(pages['domain'])
pages['created_at'] = label_encoder.fit_transform(pages['created_at'])
pages['visited_at'] = label_encoder.fit_transform(pages['visited_at'])
pages['is_frontpage'] = label_encoder.fit_transform(pages['is_frontpage'])
pages['size'] = label_encoder.fit_transform(pages['size'])
pages['path'] = label_encoder.fit_transform(pages['path'])
```

```python
#extract the necessary features that need to do the clustering
pagesNew = pages[['url', 'title', 'code', 'domain', 'created_at',
       'visited_at', 'is_frontpage', 'size', 'path'
                   ]]
```

```python
#view the encoded dataframe
pagesNew.head()
```

```python
#reduce the 30 dimensions to two dimensions using T-SNE
GT_TSNE = TSNE(n_components=2).fit_transform(pagesNew)
```

### T-SNE

T-SNE

The page data set contains lots of dimension such as( 'url', 'title', 'code', 'domain', 'created_at',
 'visited_at', 'is_frontpage', 'size', 'path'), By using the T-SNE algorithm it was redacted to 2 and project that input space to hierarchical clustering agglomerative (single, complete, ward, average). 
The reason being the HC is to identify page clusters and based on their qualities then identify the page clusters using the wallet types of the bitcoin addresses. 



```python
#reduced dimension distribution
fig, ax = plt.subplots(figsize=(7,7))
plt.scatter([i for i in GT_TSNE], [i for i in GT_TSNE])
plt.show()
```

### Hierarchical clustering agglomerative (single)

```python
#hierarchical clustering the dimension redacted data set - single
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt

linked = linkage(GT_TSNE, 'single')

labelList = range(len(GT_TSNE))

plt.figure(figsize=(20, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```
