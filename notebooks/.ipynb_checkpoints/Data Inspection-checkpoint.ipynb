##### The purpose is this notebook to give a quick inspection of the data. 


```python
%matplotlib inline
import MySQLdb as mariadb
import pandas as pd
import hashlib

from elasticsearch import Elasticsearch
from elasticsearch_dsl import Search
from elasticsearch_dsl import Q

db_host = 'd-spider-db'
db_user = 'torscraper'
db_password = 'password'
db = 'tor'
```

```python
def query_execute(query): 
    conn = mariadb.connect(db_host, db_user, db_password, db, use_unicode=True, charset="utf8");
    domain_data = pd.io.sql.read_sql(query, conn)
    conn.close()
    domain_data = domain_data.drop_duplicates(ignore_index=True)
    return domain_data
```

```python
TABLES = query_execute("SHOW TABLES;")
```

```python
TABLES
```

```python
bitcoin_address =  query_execute('select * from bitcoin_address limit 100;')
bitcoin_address_link =  query_execute('select * from bitcoin_address_link limit 100;')
category =  query_execute('select * from category limit 100;')
category_link =  query_execute('select * from category_link limit 100;')
categorylink =  query_execute('select * from categorylink limit 100;')
clone_group =  query_execute('select * from clone_group limit 100;')
daily_stat =  query_execute('select * from daily_stat limit 100;')
domain =  query_execute('select * from domain limit 100;')
email =  query_execute('select * from email limit 100;')
email_link =  query_execute('select * from email_link limit 100;')
headless_bot =  query_execute('select * from headless_bot limit 100;')
headlessbot =  query_execute('select * from headlessbot limit 100;')
open_port =  query_execute('select * from page limit 100;')
page =  query_execute('select * from category_link limit 100;')
page_link =  query_execute('select * from page_link limit 100;')
request_log =  query_execute('select * from request_log limit 100;')
search_log =  query_execute('select * from search_log limit 100;')
ssh_fingerprint =  query_execute('select * from ssh_fingerprint limit 100;')
web_component =  query_execute('select * from web_component limit 100;')
web_component_link =  query_execute('select * from web_component_link limit 100;')
```
