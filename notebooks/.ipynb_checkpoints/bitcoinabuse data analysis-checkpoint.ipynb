```python
import pandas
import json
import glob
```

```python
files = glob.glob('../data/bitcoinabuse crawler data/json/*')
```

```python
DATA = list()
err = list()
```

```python
for file in files:
    data = open(file,'r').read()
    for con in data.split('\n'):
        con  = con.replace("\'", "\"")
        con = con.replace('u"', '"')
        try:
            DATA.append(json.loads(con))
        except:
            err.append(file)
```
